```python
import operator
p=0;
letter='';
layer=1;
letters=0;
black=4*' ';
middle=list();
vari=list();
tmp=list();
flist=list();
v2=list();
v3=list();
t_time=-1;
```

```python
#判断是否是 标识符
def is_iden():
    #print(letter[0].isalpha);
    if letter[0].isalpha()==0:
        return 0;
    else:
        for i in range(1,len(letter)):
            if letter[i].isalnum==0:
                return 0;
        return 1;
#是否是常量
def is_number():
    for i in range(0,len(letter)):
        if letter[i].isdigit()==0:
            return 0;
    return 1;
def is_plus_minus():#是否为 + ，- 
    if (letter=='+') or (letter=='-'):
        return 1
    return 0
def is_time_divide():#是否为 *,/
    if (letter=='*') or (letter=='/'):
        return 1
    return 0
def is_greater_then_less_than():
    if (letter=='>') or (letter=='<'):
        return 1
    return 0
def is_guanxi():
    if (letter=='==') |(letter=='<=')|(letter=='<')|(letter=='>')|(letter=='>=')|(letter=='<>'):
        #print(234)
        return 1;
    return 0;
```

```python
#判断是不是个项
def is_factor():
    black=4*' '*layer;
    read_string();
    
    if is_iden()==1:
        tmp.append(letter);
        #print('factor',tmp)
        #print(black,"标识符 ",letter);
        read_string();
        if is_time_divide()==1:
            tmp.append(letter);
            #print(black,letter);
            is_factor();
        return 1;
    elif is_number()==1:
        tmp.append(letter);
        #print('factor',tmp)
       # print(black,"常数",letter);
        read_string();
        if is_time_divide()==1:
            tmp.append(letter);
            #print(black,letter);
            is_factor();
        return 1;
    #print('tmp end',tmp);
    else:
        return 0;
```

```python
#把list里内容合起来，并用 t0 替换
def combine(line):
    global t_time;
    t_time=t_time+1;
    char='';
    for i in range(0,3):
        if line[i] in v2:
            index_1=v2.index(line[i]);
            line[i]=v3[index_1];
        char=char+line[i];
    char1='t'+str(t_time);
    char=char1+'='+char
    middle.append(char);
    del line[0];
    del line[0];
    line[0]=char1;
    #print('combine',line)
```

```python
#判断是不是  “+” 项
def is_plus_factor():
    black = 4*' '*layer;
    if is_factor():
        #print('tmp',tmp)
        while len(tmp) > 1:
            combine(tmp);
        
        vari.append(tmp[0])
        tmp.clear();
        if is_plus_minus():
            vari.append(letter);
           # print(black,letter);
            is_plus_factor()
                #print("缺少因子");
    else:
        return 0;
            
        
```

```python
#条件判断
def is_ifconditon():
    black = 4*' '*layer;
    if is_expression():
        if is_guanxi():
            #print(black,letter);
            vari.append(letter);
            is_expression()
        else:
            print("缺少关系运算符");
```

```python
#判断是不是表达式
def is_expression():
    black=4*' '*layer;
    if is_factor():
        #print('tmp',tmp)
        while len(tmp) > 1:
            combine(tmp);
        vari.append(tmp[0])
        tmp.clear();
        if is_plus_minus():
            #print(black,letter);
            vari.append(letter);
            if is_plus_factor() ==0:
                print("缺少因子");
            
        return 1;
    
```

```python
#常量说明
def Const_description():
    read_string()
    if is_iden()==1:
        #print(black,"标识符 ",letter)
        read_string();
        if letter == '=': 
            #print(black,"赋值符号",letter)
            read_string();            
            if is_number()==1:
                #print(black,"常量",letter);
                read_string();
                if letter==',':
                   # print(black,"逗号",letter)
                    Const_description();
                elif letter==';':
                    #print(black,"分号",letter)
                    return 1;
        else:
            print("常量定义错误")
            return 0;
    else:
        print("常量定义错误");
#Var说明
def Var_description():
    read_string()
    if is_iden()==1:
        #print(black,"标识符 ",letter)
        read_string();
        if letter == ',': 
          #  print(black,"逗号",letter)
            Var_description();
        elif letter ==';':
           # print(black,"分号",letter);
            return 1;
        else:
            print("Var说明错误")
            return 0;
    else:
        print("Var说明错误");
```

```python
#语句说明
def sentense():
    global layer,letters,middle,vari;
    
    layer=layer+1;
    black=layer*4*' ';
    
    if letter == 'begin':
        
        read_string();
        while letter!='end':
            if letter=='':               
                return ;
            sentense();
            if letters==0:
                read_string();
            letters=0;
        #print(black,letter);
            
    elif letter == 'if':
        #print(black,"条件语句 if");
        is_ifconditon();
        #read_string();
        if letter == 'then':
           # print(black,"then")
            read_string();
            sentense();
            if letters==0:
                read_string();
            if letter == 'else':
               # print(black,letter);
                read_string();
                sentense();
#             else:
#                 print("缺少 else")
        else:
            print("缺少 then");
    elif letter =='while':
        #print(black,"循环语句 while");
        is_ifconditon();
        if letter=='do':
            #print(black,letter);
            read_string();
            sentense();
    elif is_iden()==1:
        flist.append(letter);
        #print(black,"标识符",letter);
        read_string();
        
        if letter=='=':
            is_expression();
            if letter ==';':
                while len(vari)>1:
                    combine(vari);
                #print('vari',vari);
                v3.append(vari[0]);
                v2.append(flist[0]);
                
                char= flist[0] +'='+vari[0];
                middle.append(char);
                flist.clear();
                vari.clear();
                ;
            else:
                print("缺少分号");
        else:
            letters=1;
            
            print("缺少等号,赋值语句错误");

    layer=layer-1;
```
